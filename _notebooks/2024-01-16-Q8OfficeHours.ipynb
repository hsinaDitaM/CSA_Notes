Consider the Following Method:

```python
public static int[] operation(int[][] matrix, int r, int c)
{
    int[] result = new int[matrix.length];
    for (int j = 0; j < matrix.length; j++)
    {
        result[j] = matrix[r][j] * matrix[j][c];
    }
    return result;
}
```
